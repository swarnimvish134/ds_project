# Google Play Store Apps: Data Cleaning and Exploratory Data Analysis

## DATA PREPARATION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/swarn/Downloads/archive (2)/googleplaystore.csv")
```

```python
df.head()
```

```python
df.info
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['Type'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.head()
```

```python
df['Installs'] = df['Installs'].str.replace('+','')
```

```python

df['Installs'] = df['Installs'].str.replace(',','')
```

```python
df.head()
```

```python
df.info
```

```python
df.dtypes
```

```python
df['Installs'].unique()
```

```python
df = df[df['Installs'] != 'Free']
```

```python
df = df[df['Installs'] != 'Free'].copy()
```

```python
df['Installs'] = df['Installs'].astype(int)
```

```python
df.dtypes
```

```python
df['Price'] = df['Price'].str.replace('$','').astype(float)
```

```python
df['Size'] = df['Size'].str.replace('M','')
```

```python
df['Size'] = pd.to_numeric(df['Size'], errors='coerce')
```

```python
df['Size'] = df['Size'].fillna(df['Size'].median())
```

```python
df.dtypes
```

## VISUALISATION

### RATING DISTRIBUTION 

```python
df['Rating'].hist()
plt.title("Rating Distribution")
plt.show()
```

''' CONCLUSUION: Most app ratings are concentrated between 4.0 and 4.5,
indicating generally positive user feedback. Very few apps have low ratings,
showing that poorly rated apps are less common.'''


### INSTALLS / CATEGORY

```python
df.groupby('Category')['Installs'].sum().plot(kind='bar')
plt.xticks(rotation=90)
plt.title("Installs by Category")
plt.show()
```

''' CONCLLUSION: Categories like Communication, Social, and Tools have the highest installs,
showing strong user demand for daily-use apps. 'Specialized categories' have lower installs, 
and totals may be influenced by the number of apps in each category.'''

### FREE VS PAID

```python
df['Type'].value_counts().plot(kind='bar')
plt.title("Free vs Paid Apps")
plt.show()
```

```python
df['Type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Free vs Paid Apps")
plt.ylabel('')  
plt.show()
```

''' CONCLUSION: MOST APPS ARE FREE '''

### Which features are related to rating?(Correlation between variables)

```python
import seaborn as sns

corr = df[['Rating','Reviews','Installs','Size','Price']].corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

''' CONCLUSION: Ratings show very weak correlation with all features,
meaning they are not strongly influenced by installs, reviews, size, or price.
The only notable relationship is a moderate positive correlation between reviews and installs (0.63).
 '''

### INSTALL VS RATING

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(df['Installs'], df['Rating'])
plt.xlabel("Installs")
plt.ylabel("Rating")
plt.title("Installs vs Rating")
plt.show()
```

''' conclusion: There’s no strong correlation between installs and ratings—apps
with both low and high install counts tend to have similar rating distributions.
Most ratings cluster between 4.0 and 4.5 regardless of install size.
'''
